### 3. The .str.replace() Method

```python
chicago = pd.read_csv("../00_Datasers/chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
"Hello world".replace("l", "!")
```

```python
chicago["Department"] = chicago["Department"].str.replace("MGMNT", "MANAGEMENT")
chicago.head()
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$", "").astype(float)
```

```python
chicago["Employee Annual Salary"].sum()
```

### 4. Filtering with String Method

```python
chicago = pd.read_csv("../00_Datasers/chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
mask = chicago["Position Title"].str.lower().str.contains("water")
```

```python
chicago[mask]
```

```python
mask = chicago["Position Title"].str.lower().str.startswith("water")
```

```python
chicago[mask]
```

```python
mask = chicago["Position Title"].str.lower().str.endswith("ith")
```

```python
chicago[mask]
```

### 5. More String Methods - .strip(), .lstrip() and .rstrip()

```python
chicago = pd.read_csv("../00_Datasers/chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
"      Hello World     ".lstrip()
```

```python
"      Hello World     ".strip()
```

```python
chicago["Name"].str.lstrip()
```

```python
chicago["Name"].str.rstrip()
```

```python
chicago["Name"] = chicago["Name"].str.lstrip().str.rstrip()
```

```python
chicago["Position Title"] = chicago["Position Title"].str.strip()
```

### 6. String Methods on Index and Columns

```python
chicago = pd.read_csv("../00_Datasers/chicago.csv", index_col="Name").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago.index = chicago.index.str.strip().str.title()
```

```python
chicago.head(3)
```

```python
chicago.columns = chicago.columns.str.upper()
```

```python
chicago.head(3)
```

### 9. The expand and n Parameters of the str.split() Method

```python
chicago = pd.read_csv("../00_Datasers/chicago.csv").dropna(how = "all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head(3)
```

```python
chicago[ ["First Name", "Last Name"] ] = chicago["Name"].str.split(",", expand=True)
```

```python
chicago.head(3)
```

```python
chicago[["First Title Word", "Remaining Words"]] = chicago["Position Title"].str.split(" ", expand = True, n = 1)
```

```python
chicago.head(3)
```
